```python
# =================================================================
# PROJETO MISSION AURORA SIGER - SISTEMA DE VERIFICAÇÃO DE VOO
# =================================================================

# === 1. DADOS BASE (PARAMETRIZAÇÃO) ===
capacidade_total_kwh = 500  # Capacidade total da bateria
taxa_perda_percent = 8      # Perda por eficiência térmica/transmissão
consumo_decolagem_kwh = 300 # Energia necessária para a ignição

# Limites de Segurança (Setpoints)
TEMP_INT_MIN, TEMP_INT_MAX = 15, 35
TEMP_EXT_MIN, TEMP_EXT_MAX = -50, 60
PRESSAO_MIN, PRESSAO_MAX = 120, 150
CARGA_MIN_PERCENT = 80

# === 2. ENTRADA DE DADOS (TELEMETRIA) ===
print("--- ENTRADA DE DADOS DE TELEMETRIA ---")
perc_atual = float(input("Porcentagem atual da carga (%): "))
temp_interna = float(input("Temperatura interna (°C): "))
temp_externa = float(input("Temperatura externa (°C): "))
pressao = float(input("Pressão do tanque (bar): "))
integridade_estrutural = int(input("Integridade estrutural (1-OK / 0-FALHA): "))

# Módulos Críticos
print("\nInforme o status dos módulos (OK ou FALHA):")
mod_nav = input("Módulo de Navegação: ").strip().upper()
mod_com = input("Módulo de Comunicação: ").strip().upper()
mod_prop = input("Módulo de Propulsão: ").strip().upper()

# === 3. PROCESSAMENTO E CÁLCULOS (ANÁLISE ENERGÉTICA) ===
# Cálculo baseado na relação Energia = Potência x Tempo e Eficiência
energia_disponivel = capacidade_total_kwh * (perc_atual / 100)
perda_energia = energia_disponivel * (taxa_perda_percent / 100)
energia_util = energia_disponivel - perda_energia
autonomia_restante = energia_util - consumo_decolagem_kwh

# === 4. ALGORITMO DE VERIFICAÇÃO (LÓGICA BOOLEANA) ===
# Verificação de faixas seguras
temp_safe = TEMP_INT_MIN <= temp_interna <= TEMP_INT_MAX
temp_ext_safe = TEMP_EXT_MIN <= temp_externa <= TEMP_EXT_MAX
pressao_safe = PRESSAO_MIN <= pressao <= PRESSAO_MAX
carga_safe = perc_atual >= CARGA_MIN_PERCENT
energia_suficiente = energia_util >= consumo_decolagem_kwh
integridade_ok = integridade_estrutural == 1

# Verificação dos módulos
nav_ok = mod_nav == "OK"
com_ok = mod_com == "OK"
prop_ok = mod_prop == "OK"
sistemas_criticos_ok = nav_ok and com_ok and prop_ok

# === 5. ANÁLISE ASSISTIDA POR IA (IDENTIFICAÇÃO DE ANOMALIAS) ===
alertas_risco = []
if 32 <= temp_interna <= 35:
    alertas_risco.append("ALERTA: Temperatura interna próxima ao limite máximo (Risco de fadiga).")
if perc_atual < 85 and perc_atual >= 80:
    alertas_risco.append("CUIDADO: Margem de energia próxima ao limite mínimo de segurança.")

# === 6. IMPRESSÃO DOS RESULTADOS ===
print("\n" + "="*30)
print("   RELATÓRIO FINAL DE VOO")
print("="*30)

# Exibição de Status Detalhado
print(f"Energia Útil: {energia_util:.2f} kWh")
print(f"Sobra após Decolagem: {autonomia_restante:.2f} kWh")
print(f"Status Estrutural: {'ÍNTEGRO' if integridade_ok else 'FALHA'}")

print("\n--- CHECKLIST DE SISTEMAS ---")
print(f"[ {'OK' if pressao_safe else '!!' } ] Pressão dos Tanques")
print(f"[ {'OK' if temp_safe else '!!' } ] Temperatura Interna")
print(f"[ {'OK' if carga_safe else '!!' } ] Nível de Carga")
print(f"[ {'OK' if sistemas_criticos_ok else '!!' } ] Módulos Críticos")

# Exibição de Alertas de Anomalias
if alertas_risco:
    print("\n--- ANÁLISE DE RISCO ---")
    for aviso in alertas_risco:
        print(aviso)

# VEREDITO FINAL
print("\n" + "="*30)
if (pressao_safe and temp_safe and temp_ext_safe and
    carga_safe and energia_suficiente and
    integridade_ok and sistemas_criticos_ok):
    print("STATUS: !!! PRONTO PARA DECOLAR !!!")
else:
    print("STATUS: !!! DECOLAGEM ABORTADA !!!")
    print("MOTIVO: Verifique as falhas apontadas no checklist acima.")
print("="*30)
```
