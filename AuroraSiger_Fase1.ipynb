```python
# === DADOS BASE ===
capacidade_total = 500
taxa_perda = 8
consumo_decolagem = 300
energia_operacional = 320

temp_interna_min = 15
temp_interna_max = 35

temp_externa_min = -50
temp_externa_max = 60

pressao_min = 120
pressao_max = 150

carga_minima_percent = 80
```

```python
# === INPUTS ===
perc_atual = int(input("Porcentagem atual da carga(%):...."))
temp_interna = int(input("Temperatura interna(°C):....."))
temp_externa = int(input("Temperatura externa(°C):....."))
pressao = int(input("Pressão do tanque(kPa):....."))
integridade_estrutural =  int(input("Integridade estrutural(0-FALHA/1-OK):.... "))
modulo_navegacao = input("Modulo de navegação(OK/FALHA):....")
modulo_comunicacao = input("Modulo de comunicação(OK/FALHA):....")
modulo_propulsao = input("Modulo de propulsao(OK/FALHA):....")
```

```python
# === VERIFICAÇÃO DA INTEGRIDADE ESTRUTURAL ===
if integridade_estrutural == 0:
    integridade_estrutural = False
elif integridade_estrutural == 1:
    integridade_estrutural = True
```

```python
# === VERIFICAÇÃO DOS MÓDULOS ===
if modulo_propulsao.upper() == "OK":
  modulo_propulsao = True
else:
  modulo_propulsao = False

if modulo_comunicacao.upper() == "OK":
  modulo_comunicacao = True
else:
  modulo_comunicacao = False

if modulo_navegacao.upper() == "OK":
  modulo_navegacao = True
else:
  modulo_navegacao = False
```

```python
# === CÁLCULOS ===
energia_disponivel = capacidade_total * (perc_atual/100)
perda_energia = energia_disponivel * (taxa_perda/100)
energia_util = energia_disponivel - perda_energia
margem_energia = energia_util - consumo_decolagem
margem_percent = (margem_energia/energia_util) * 100

if energia_util >= consumo_decolagem:
    energia_safe = True
else:
    energia_safe = False

if temp_interna_min <= temp_interna <= temp_interna_max:
    temp_safe = True
else:
    temp_safe = False

if temp_externa_min <= temp_externa <= temp_externa_max:
    temp_externa_safe = True
else:
    temp_externa_safe = False

if pressao_min <= pressao <= pressao_max:
    pressao_safe = True
else:
    pressao_safe = False

if perc_atual >= carga_minima_percent:
    carga_safe = True
else:
    carga_safe = False

```

```python
# === IMPRIMINDO RESULTADOS ===
print("==========VERIFICANDO==========")
if pressao_safe:
  print("Pressão do Tanque (kPa): OK")
else:
  print("Pressão do tanque (kPa): ERRO")
if temp_safe:
  print("Temperatura (°C): OK")
elif temp_safe >= 32:
  print("Temperatura (°C): ALTA | TEMPERATURA MUITO PRÓXIMA DO LIMITE ESPERADO | RISCO: PERDA DE FADIGA/DESGASTE DE MATERIAL")
else:
  print("Temperatura (°C): ERRO")
if modulo_propulsao:
  print("Propulsao: OK")
else:
  print("Propulsão: ERRO")
if modulo_comunicacao:
  print("Comunicação: OK")
else:
  print("Comunicação: ERRO")
if modulo_navegacao:
  print("Navegação: OK")
else:
  print("Navegação: ERRO")
if pressao_safe and temp_safe and modulo_navegacao and modulo_propulsao and modulo_comunicacao and energia_safe:
  print("!!!PRONTO PARA DECOLAR!!!")
else:
  print("!!!DECOLAGEM ABORTADA!!!")
```
